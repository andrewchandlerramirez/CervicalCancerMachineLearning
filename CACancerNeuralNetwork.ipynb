```python
import numpy as np


#taking training and test file, then seperating them into inputs and outputs
data_file = open("sobar-72.csv", 'r')
data_list = data_file.readlines()
data_file.close()


xdata = []
ydata = []
for line in data_list[1:]:
    values = line.split(',')
    xdata.append([float(x) for x in values[0:len(values)-1]])
    index = int(values[len(values)-1])
    array = [0,0] #numpy.zeros(10)
    array[index] = 1
    ydata.append(array)


```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(xdata,ydata,test_size= 0.2)
```

```python
import tensorflow as tf


xtrain = tf.keras.utils.normalize(xtrain, axis = 1)
xtest = tf.keras.utils.normalize(xtest, axis = 1)
 

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units = 128, activation = 'relu', input_dim = 19))
model.add(tf.keras.layers.Dense(units = 128, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 128, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 64, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 32, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 2, activation = 'sigmoid'))



model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

xtrain = np.asarray(xtrain)
ytrain = np.asarray(ytrain)
xtest = np.asarray(xtest)
ytest = np.asarray(ytest)


model.fit(xtrain, ytrain, epochs=200)
model.evaluate(xtest, ytest)
```

```python
model.save('cc.model')
```
