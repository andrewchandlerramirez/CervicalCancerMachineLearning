```python
import numpy as np


#taking training and test file, then seperating them into inputs and outputs
data_file = open("sobar-72.csv", 'r')
data_list = data_file.readlines()
data_file.close()


xdata = []
ydata = []
for line in data_list[1:]:
    values = line.split(',')
    xdata.append([float(x) for x in values[0:len(values)-1]])
    ydata.append(int(values[len(values)-1]))

```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(xdata,ydata,test_size= 0.2)
```

```python
#F
from sklearn import tree
DecisionTreeModel = tree.DecisionTreeClassifier(criterion='entropy')
#G
DecisionTreeModel.fit(xtrain,ytrain)
```

```python
print(DecisionTreeModel.score(xtest,ytest))
```

```python
names = data_list[0].split(',')
features = names[:len(names)-1]
targets = [names[len(names)-1]]
for i in features:
    print(i)
for i in targets:
    print(i)

print(len(features),len(targets),len(xtest[0]))
```

```python
from matplotlib import pyplot as plt
#H
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(DecisionTreeModel,
                    feature_names= features,
                    class_names= targets[0],
                    filled = True,
                    fontsize = 12)
```
